```python
import numpy as np
import pygplates
import pygmt

%load_ext autoreload
%autoreload 2
```

```python
euler_pole_position = (60,10)
euler_pole_angle = 60.

continent_A = pygplates.PolygonOnSphere([(50,-50),(58,-56),(60,-80),(60,-120),(40,-130),(20,-105),(30,-80),(35,-63)])   
pole_of_rotation = pygplates.FiniteRotation(pygplates.PointOnSphere(euler_pole_position), np.radians(euler_pole_angle))
recontructed_polygon = pole_of_rotation * continent_A
```

```python
fig = pygmt.Figure()
fig.coast(projection="G-70/52/12c", region="g", frame="g", land="gray")

fig.plot(x=continent_A.get_points().to_lat_lon_array()[:,1], y=continent_A.get_points().to_lat_lon_array()[:,0], color='red')

fig.plot(x=recontructed_polygon.get_points().to_lat_lon_array()[:,1], y=recontructed_polygon.get_points().to_lat_lon_array()[:,0], color='blue')

fig.plot(x=euler_pole_position[1], y=euler_pole_position[0], style='a0.8c', color='black')

fig.show()
```

```python
fig = pygmt.Figure()
fig.basemap(projection="G-70/52/12c", region="g", frame="g")

fig.plot(x=continent_A.get_points().to_lat_lon_array()[:,1], y=continent_A.get_points().to_lat_lon_array()[:,0], color='red')

fig.plot(x=recontructed_polygon.get_points().to_lat_lon_array()[:,1], y=recontructed_polygon.get_points().to_lat_lon_array()[:,0], color='blue')

for interpolated_pole_angle in np.linspace(0,euler_pole_angle,10): 
    pole_of_rotation = pygplates.FiniteRotation(pygplates.PointOnSphere(euler_pole_position), np.radians(interpolated_pole_angle))
    recontructed_polygon = pole_of_rotation * continent_A
    fig.plot(x=recontructed_polygon.get_points().to_lat_lon_array()[:,1], y=recontructed_polygon.get_points().to_lat_lon_array()[:,0], pen='0.5p,grey', close=True)

fig.plot(x=euler_pole_position[1], y=euler_pole_position[0], style='a0.8c', color='black')

fig.show()
```

```python
for interpolated_pole_angle in np.linspace(0,euler_pole_angle,11): 

    fig = pygmt.Figure()
    fig.basemap(projection="G-70/52/12c", region="g", frame="g")

    fig.plot(x=continent_A.get_points().to_lat_lon_array()[:,1], y=continent_A.get_points().to_lat_lon_array()[:,0], color='red')

    pole_of_rotation = pygplates.FiniteRotation(pygplates.PointOnSphere(euler_pole_position), np.radians(interpolated_pole_angle))
    recontructed_polygon = pole_of_rotation * continent_A
    #fig.plot(x=recontructed_polygon.get_points().to_lat_lon_array()[:,1], y=recontructed_polygon.get_points().to_lat_lon_array()[:,0], pen='0.5p,grey', close=True)

    fig.plot(x=recontructed_polygon.get_points().to_lat_lon_array()[:,1], 
             y=recontructed_polygon.get_points().to_lat_lon_array()[:,0], color='blue')

    fig.plot(x=[euler_pole_position[1], recontructed_polygon.get_interior_centroid().to_lat_lon()[1]],
             y=[euler_pole_position[0], recontructed_polygon.get_interior_centroid().to_lat_lon()[0]], pen='1p,black')

    fig.plot(x=[euler_pole_position[1], continent_A.get_interior_centroid().to_lat_lon()[1]],
             y=[euler_pole_position[0], continent_A.get_interior_centroid().to_lat_lon()[0]], pen='1p,black')

    fig.plot(x=euler_pole_position[1], y=euler_pole_position[0], style='a0.8c', color='black')

    fig.savefig('plots/continent_motion_{:0.0f}_degrees.png'.format(interpolated_pole_angle))
    
```

```python
fig = pygmt.Figure()
fig.coast(projection="G-70/52/12c", region="g", frame="g", land="gray")

fig.plot(x=continent_A.get_points().to_lat_lon_array()[:,1], y=continent_A.get_points().to_lat_lon_array()[:,0], color='red')

fig.plot(x=recontructed_polygon.get_points().to_lat_lon_array()[:,1], y=recontructed_polygon.get_points().to_lat_lon_array()[:,0], color='blue')

fig.plot(x=euler_pole_position[1], y=euler_pole_position[0], style='a0.8c', color='black')

small_circle_diameters = np.arange(7000, 20000, 1000)

fig.plot(x=np.ones_like(small_circle_diameters)*euler_pole_position[1],
         y=np.ones_like(small_circle_diameters)*euler_pole_position[0],
         sizes = small_circle_diameters,
         style='E-',
         pen='2p,darkblue',
         transparency=80)

fig.show()
```
