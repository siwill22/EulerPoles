```python
import numpy as np
import pygplates
import pygmt

%load_ext autoreload
%autoreload 2
```

```python
polygons = pygplates.FeatureCollection('./SouthAtlantic/SouthAtlanticCoastlines.shp')
rotations = pygplates.RotationModel('./SouthAtlantic/Global_EarthByte_230-0Ma_GK07_AREPS.rot')

SAM_polygon = polygons.get(lambda feature: feature.get_reconstruction_plate_id() == 201).get_geometry()
AFR_polygon = polygons.get(lambda feature: feature.get_reconstruction_plate_id() == 701).get_geometry()

print(SAM_polygon)
```

```python
fig = pygmt.Figure()
fig.coast(projection="G-20/5/12c", region="g", frame="g", land="gray")

fig.plot(x=AFR_polygon.get_points().to_lat_lon_array()[:,1], 
         y=AFR_polygon.get_points().to_lat_lon_array()[:,0], color='red')

reconstruction_time = 150.
    
pole_of_rotation = rotations.get_rotation(reconstruction_time, 201, from_time=0, fixed_plate_id=701)
euler_pole_position = pole_of_rotation.get_lat_lon_euler_pole_and_angle_degrees()
#print(pole_of_rotation)

SAM_reconstructed = pole_of_rotation * SAM_polygon

fig.plot(x=SAM_reconstructed.get_points().to_lat_lon_array()[:,1], 
         y=SAM_reconstructed.get_points().to_lat_lon_array()[:,0], 
         pen='2p,darkblue', color='blue', transparency=50, close=True)

fig.plot(x=euler_pole_position[1], y=euler_pole_position[0], style='a0.8c', color='black')

fig.show()
```

```python
fig = pygmt.Figure()
fig.coast(projection="G-20/5/12c", region="g", frame="g", land="gray")

fig.plot(x=AFR_polygon.get_points().to_lat_lon_array()[:,1], 
         y=AFR_polygon.get_points().to_lat_lon_array()[:,0], color='red')

for reconstruction_time in np.arange(0,150,20):
    
    pole_of_rotation = rotations.get_rotation(reconstruction_time, 201, from_time=0, fixed_plate_id=701)
    euler_pole_position = pole_of_rotation.get_lat_lon_euler_pole_and_angle_degrees()
    #print(pole_of_rotation)
    
    SAM_reconstructed = pole_of_rotation * SAM_polygon

    fig.plot(x=SAM_reconstructed.get_points().to_lat_lon_array()[:,1], 
             y=SAM_reconstructed.get_points().to_lat_lon_array()[:,0], 
             pen='0.8,blue', transparency=50, close=True)

    fig.plot(x=euler_pole_position[1], y=euler_pole_position[0], style='a0.8c', color='black')

fig.show()
    
```
